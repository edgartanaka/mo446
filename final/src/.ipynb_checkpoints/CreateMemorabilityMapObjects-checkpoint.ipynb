```python
test_features = target_features['sptHistObjects'][rtr['test_indices']]
train_features = target_features['sptHistObjects'][rtr['train_indices']]
```

```python
object_scores = np.empty([len(areas)+1,num_images],dtype=float)
object_scores[:] = np.nan
```

```python
n_obj_classes = test_features.shape[1]//5
n_obj_classes
```

```python
def CalcObjectExclusionFactorSptHistObjs(train_features, test_features, prediction1, kernel_param, model, i_out):  
    n_obj_classes = len(test_features)//5 # test features is 2-level sptHist of object class data

    # remove feature at index i_out
    replace_with = 0
    test_features[i_out] = replace_with
    test_features[i_out+n_obj_classes] = replace_with
    test_features[i_out+(2*n_obj_classes)] = replace_with
    test_features[i_out+(3*n_obj_classes)] = replace_with
    test_features[i_out+(4*n_obj_classes)] = replace_with
    
    test_features = np.expand_dims(test_features, axis=0)
    # make new prediction without feature at index i_out
    K_test = compute_kernel(test_features, train_features,kernel_param)
#     K_test = np.concatenate([np.expand_dims(np.arange(1,test_features.shape[0]+1), axis=0),K_test],axis=1)
    
    prediction2 = model.predict(K_test)

    exclusion_factor = prediction1-prediction2

    return exclusion_factor
```

```python
for i in range(num_images):
    curr_test_features = test_features[i].copy()
    which_objects = np.where(curr_test_features[:n_obj_classes] != 0)[0]
    for j in range(len(which_objects)):
        object_scores[which_objects[j]+1,i] = CalcObjectExclusionFactorSptHistObjs(train_features, curr_test_features, predicted[i], params, model, which_objects[j])
        
```

```python
min_area = 4000

object_presences = np.array(areas[:,rtr['test_indices']] > min_area,dtype=float)
object_presences = np.vstack([np.zeros((1,object_presences.shape[1])), object_presences]) # first position is for 0/'none' label (refers to unlabeled segments of image
```

```python
rank_object_scores = np.empty((len(areas)+1,num_images))
rank_object_scores[:] = np.nan

i1 = predicted.argsort()[::-1]
i2 = i1.argsort()
```

```python
for i in range(num_images):
    
    which_objects = np.argwhere(~np.isnan(object_scores[:,i]))
    for j in range(len(which_objects)):
        
        a = which_objects[j]
        score_for_object_a_in_image_i = object_scores[a,i]
        
        predicted2 = predicted
        predicted2[i] += score_for_object_a_in_image_i # set of predictions when object a is omitted from image i

        j1 = predicted2.argsort()[::-1]
        j2 = j1.argsort()
        
        new_rank_for_image_i = j2[i]
        old_rank_for_image_i = i2[i]
        
        rank_change_for_image_i = old_rank_for_image_i-new_rank_for_image_i
        
        rank_object_scores[a,i] = rank_change_for_image_i
```

```python
avg_object_scores = np.empty((object_scores.shape[0],1))
avg_object_scores[:] = np.nan #nans are set to zero for sorting below (zero since nan means we have no data so should go with prior guess of 0 effect)

avg_rank_object_scores = np.empty((rank_object_scores.shape[0],1))
avg_rank_object_scores[:] = np.nan
```

```python
for i in range(len(avg_object_scores)):
    
    curr_object_scores = object_scores[i,:]
    curr_rank_object_scores = rank_object_scores[i,:]

    #which_scores = find(~isnan(curr_object_scores));
    which_scores = np.where(object_presences[i,:])[0]
    if which_scores.size != 0:
        avg_object_scores[i] = np.mean(curr_object_scores[which_scores])
        avg_rank_object_scores[i] = np.mean(curr_rank_object_scores[which_scores])
```

```python
avg_object_scores= np.nan_to_num(avg_object_scores) # remove nans
avg_rank_object_scores = np.nan_to_num(avg_rank_object_scores)
```

```python
def MontageSubplots(ims):
    N = ims.shape[3]
    nc = int(np.floor(np.sqrt(N)))
    nr = int(np.ceil(N/nc))
    fig, axes = plt.subplots(figsize=(8,8),nrows=nr, ncols=nc)
    
    for i, ax in enumerate(axes.flatten()):
        ax.imshow(ims[:,:,:,i])
        ax.axis('off')    
    plt.show()
```

```python
def VisualizeOnePredictedSet(predct, test_label, test_set_img, which_set, k):
    N = test_label.shape[0]
    i = predct.argsort(axis=0)[::-1]

    if which_set == 'top':
        r = range(k)
        # plot
        MontageSubplots(test_set_img[:,:,:,i[r]])
        curr_actual = test_label[i[r]]
        print('Top memorability\nactual mean: {}, min: {}, max: {}'.format(np.mean(curr_actual), min(curr_actual), max(curr_actual)))
        
    elif which_set == 'middle':
        r = range(int(np.floor(N/2-k/2)+1),int(np.floor(N/2+k/2)) + 1)
        MontageSubplots(test_set_img[:,:,:,i[r]])
        curr_actual = test_label[i[r]]
        print('Middle memorability\nactual mean: {}, min: {}, max: {}'.format(np.mean(curr_actual), min(curr_actual), max(curr_actual)))
        
    elif which_set == 'bottom':
        r = range(N-k,N)
        MontageSubplots(test_set_img[:,:,:,i[r]])
        curr_actual = test_label[i[r]]
        print('Bottom memorability\nactual mean: {}, min: {}, max: {}'.format(np.mean(curr_actual), min(curr_actual), max(curr_actual)))
```
